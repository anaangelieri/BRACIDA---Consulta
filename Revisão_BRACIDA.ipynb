```python
# Diferentes estilos de Plots

# Dados
xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

# Subplots
fig, axes = plt.subplots(1, 4, figsize = (12,3))

# Plot 1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = "black")
axes[0].set_title("scatter")

# Plot 2
axes[1].step(n, n ** 2, lw = 2, color = "blue")
axes[1].set_title("step")

# Plot 3
axes[2].bar(n, n ** 2, align = "center", width = 0.5, alpha = 0.5, color = "magenta")
axes[2].set_title("bar")

# Plot 4
axes[3].fill_between(x, x ** 2, x ** 3, alpha = 0.5, color = "green");
axes[3].set_title("fill_between");
```

```python
# Dados simples
dados = [1, 2, 2, 3, 3, 3, 4, 5]

# Criar subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Histograma normal
axes[0].hist(dados, bins=range(1, 7), edgecolor='black', color='skyblue', align='left')
axes[0].set_title("Histograma Normal")
axes[0].set_xlabel("Valor")
axes[0].set_ylabel("Frequência")

# Histograma acumulativo
axes[1].hist(dados, bins=range(1, 7), cumulative=True, edgecolor='black', color='lightgreen', align='left')
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlabel("Valor")
axes[1].set_ylabel("Frequência Acumulada")

# Layout bonitinho
plt.tight_layout()
plt.show()
```
